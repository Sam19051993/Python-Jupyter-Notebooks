```python
import pandas as pd
input_data = pd.read_json("supermarkets.json")
input_data.set_index("Address")#set_index is not in-place. Hence, this data frame is not updated. A temporary data frame is
                                # created.
```

```python
input_data #input_data is not modified since the method set_index is not in-place
```

```python
temp_data = input_data
```

```python
temp_data.set_index("Address",inplace = True)
```

```python
temp_data #temp_data is modified since we specified explicitly that we wanted an in-place operation
```
