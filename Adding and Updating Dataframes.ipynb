```python
import pandas as pd
input_data = pd.read_csv("supermarkets.csv")
```

```python
input_data.set_index("Address",inplace = True)
input_data
```

```python
input_data.shape
```

```python
input_data["Continent"] = input_data.shape[0] * ["North America"] # Adding a new Column. In-place by default
```

```python
input_data
```

```python
input_data["Continent"] = input_data["Country"] + "," + input_data["Continent"] #Updating the Continent Column
```

```python
input_data
```

```python
#Adding a New Row: 1. Transpose the original Data Frame in a new Variable
#                  2. Add a New Column to the Transposed DataFrame
#                  3. Transpose back to the original Data Frame
```

```python
input_data_t = input_data.T # Transpose the Data Frame
```

```python
input_data_t
```

```python
input_data_t["871 Park Street"] = ["7","San Francisco","CA 94112","USA","My Shop","10","USA,North America"] # Adding a new Column
```
